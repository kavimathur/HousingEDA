### Skew and Kurtosis
Skew is a measure of asymmetry defined as the average cubed deviation from the mean. A value greater than zero implies positive skew. As can also be seen in the distribution plot above, positive skew gives us a higher frequency of values in the lower half of the total range of Sale Price (i.e. most houses in this dataset sell below ~$400k)

Kurtosis describes the degree to which the distribution peaks. A normal distribution has a kurtosis of 3, and greater than 3 defines a leptokurtic distibution. This means that there are fatter tails and a higher likelihood of extreme values for the variable in question.

```python
print "Skewness: %f" % df_train['SalePrice'].skew()
print "Kurtosis: %f" % df_train['SalePrice'].kurt()
```

## Dropping Null Values

Many of the variables have missing observations, with up to 99% null values. I'm dropping these variables altogether

```python
total = df_train.isnull().sum().sort_values(ascending = False)
percent = (df_train.isnull().sum()/df_train.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing_data.head(20)
```

```python
df_train = df_train.drop(missing_data[missing_data['Total'] > 1].index, 1)
df_train = df_train.drop(df_train.loc[df_train['Electrical'].isnull()].index) #these lines should take out all variables with missing data --> only one row for 'Electrical' taken out though, not whole var
df_train.isnull().sum().max() #confirm all NaN data cells taken out
```

## Regressions
Usually not included in EDA's, but I thought it would be useful to see the coefficients and significance for a handful of variables when regressed on Sale Price before diving into more complicated machine learning algorithms.

```python
reg = smf.ols('SalePrice ~ TotalBsmtSF + GrLivArea + YearBuilt', data=df_train).fit()
reg.summary()
```

```python
df_train['LSalePrice'] = np.log(df_train['SalePrice'])
df_train['Age'] = df_train['YrSold'] - df_train['YearBuilt']
```

```python
reg = smf.ols('LSalePrice ~ TotalBsmtSF + GrLivArea + Age', data=df_train).fit()
reg.summary()
```

## Variable for before or after housing market crash 
Creating a new variable to represent houses sold before or after the beginning of the housing crash, which I'm identifying as starting during May of 2007.

Idea: instead of using this dummy variable, it might be worthwhile to pull in some public data on housing market indices. Though I would have to use differenced data, since contemporary housing index data would not be available when forecasting.

```python
def housingcrash(time):
    if time['YrSold'] < 2007:
        val = 0
    if time['YrSold'] == 2007:
        if time['MoSold'] < 5:
            val = 0
        if time['MoSold'] >= 5:
            val = 1
    elif time['YrSold'] > 2007:
        val = 1
    return val
```

```python
df_train['Crash'] = df_train.apply(housingcrash, axis=1)
```
